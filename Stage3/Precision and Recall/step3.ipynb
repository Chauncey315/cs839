```python
import pandas as pd
table_a="Table A.xls"
table_b="Table B.xls"
candidate_set="Candidate set.xls"
dfa = pd.read_csv(table_a)
dfb = pd.read_csv(table_b)
dfc = pd.read_csv(candidate_set)
#dfp = pd.read_csv(prediction_set)
s = dfc.sample(50)
s = pd.merge(s,dfa,how='left',left_on="A_id",right_on="_id")
s = pd.merge(s,dfb,how='left',left_on="B_id",right_on="_id")
display(s)
```

```python
table = pd.merge(dfc,dfa,how='left',left_on="A_id",right_on="_id")
table = pd.merge(table,dfb,how='left',left_on="B_id",right_on="_id")
filtered_table = table[table['time_x'].str[-2:]==table['time_y'].str[-2:]]
#print (filtered_talbe.count())
#display(filtered_talbe)
s = filtered_table.sample(50)
display(s)
```
