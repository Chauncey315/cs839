```python
import pandas as pd
table_a="Table A.xls"
table_b="Table B.xls"
candidate_set="Candidate set.xls"
dfa = pd.read_csv(table_a)
dfb = pd.read_csv(table_b)
dfc = pd.read_csv(candidate_set)
#dfp = pd.read_csv(prediction_set)
s = dfc.sample(50)
s = pd.merge(s,dfa,how='left',left_on="A_id",right_on="_id")
s = pd.merge(s,dfb,how='left',left_on="B_id",right_on="_id")
display(s)
```

```python
def Jaccard(row):
    a = row['title_x']
    b = row['title_y']
    a='##'+a+'##'
    b='##'+b+'##'
    set_a=[]
    set_b=[]
    for char in range(0,len(a)-2):
        set_a.append(a[char:char+3])
    for char in range(0,len(b)-2):
        set_b.append(b[char:char+3])
    overlap=0
    for candidate in set_a:
        if candidate in set_b:
            overlap+=1
    if overlap/(len(a)+len(b)-overlap)>0.15:
        return True
    else:
        return False
    
def year(row):
    t1 = str(row['time_x'])[-2:]
    t2 = str(row['time_y'])[-2:]
    if not t1.isdigit() or not t2.isdigit():
        return True
    if abs(int(t1)-int(t2))<=1:
        return True
    else:
        return False
```

```python
table = pd.merge(dfc,dfa,how='left',left_on="A_id",right_on="_id")
table = pd.merge(table,dfb,how='left',left_on="B_id",right_on="_id")
#filtered_table = table[table['time_x'].str[-2:]==table['time_y'].str[-2:]]
filtered_table = table[table.apply(year, axis=1)]
filtered_table = filtered_table[filtered_table.apply(Jaccard, axis=1)]
print (filtered_table.count())
s = filtered_table.sample(50)
display(s)
```
